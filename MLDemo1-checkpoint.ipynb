```python
import pandas as pd
import random

# Generate random dataset
data = []
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

for i in range(1, 51):
    row = [
        i,
        round(random.uniform(4.0, 7.5), 1),  # Sepal Length
        round(random.uniform(2.0, 4.5), 1),  # Sepal Width
        round(random.uniform(1.0, 6.5), 1),  # Petal Length
        round(random.uniform(0.1, 2.5), 1),  # Petal Width
        random.choice(species)               # Random species
    ]
    data.append(row)

# Create DataFrame
df = pd.DataFrame(data, columns=['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# Save to CSV
df.to_csv("iris.csv", index=False)

print("Random dataset saved as iris.csv")
```

```python
#Loading the Dataset and Displaying the First Few Rows
iris_data = pd.read_csv('iris.csv')
iris_data.head()
```

```python
# split the data into Features (x) and labels (y)
x = iris_data.drop(columns=['Id', 'Species'])
y= iris_data['Species']
```

```python
x.head()
```

```python
# Create a ML model
model = LogisticRegression()
```

```python
# Train the model
model.fit(x.values,y)
```

```python
#predict using the trained model
predictions = model.predict([[4.5,3.5,5.9,1.9]])
```

```python
# print the predictions
print(predictions)
```
